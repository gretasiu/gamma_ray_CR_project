```python
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
import math
```

Solving the follow equation in steady state <br>

$$\frac{dn}{dt}=\frac{1}{x^2}\frac{d}{dx}(f(x,y)x^2\frac{dn}{dx})$$<br>

1.1  $f(x,y)=1$ <br>
$$0=\frac{1}{x^2}(x^2\frac{d^2n}{dx^2}+2x\frac{dn}{dx})$$ <br>
$$0=x^2\frac{d^2n}{dx^2}+2x\frac{dn}{dx}$$ <br>
let $z=\frac{dn}{dx}$
$$\frac{dz}{dx}=-\frac{2}{x}z$$

```python
# Task 1.1: f(x,y) = 1, which is a constant.
def dU_dx(U, x):
    # Here U is a vector such that n=U[0] and z=U[1]. This function hence returns [n', z']
    return [U[1], (-0.4/x)*U[1]]

#n(0)=0 and n'(0)=0
U0 = [0, 1]
#U0 = [10**-9, 10**-9]
xs = np.linspace(0.0001, 50, 200)
Us = odeint(dU_dx, U0, xs)
ns = Us[:,0]
zs = Us[:,1]

#plotting the solution. 
plt.xlabel("x")
plt.ylabel("n")
plt.plot(xs, ns)
plt.show()

plt.xlabel("x")
plt.ylabel("n'")
plt.plot(xs, zs)
plt.show()
```

```python
#Task 2.1: a problem with well known solution - damped harmonic oscillator (smaller damping)
def dU_dx(U, x):
    # Here U is a vector such that y=U[0] and z=U[1]. This function should return [y', z']
    return [U[1], -0.2*U[1] - 1*U[0]]
U0 = [1, 0]
xs = np.linspace(0, 20, 200)
Us = odeint(dU_dx, U0, xs)
ys = Us[:,0]

plt.xlabel("t")
plt.ylabel("x")
plt.title("Damped harmonic oscillator")
plt.plot(xs,ys);
```

### Part 2 -- introducing a more physical diffusion coefficeint <br>

Diffusion coefficient D(r, E) depends on the location within a molecular cloud r and the energy of the CR particle E. 

(a) $D_0 = D(r) = \frac{D_0}{(1+\frac{r}{r_c})^2}$, where $r_c$ is the charateristic MC radius and $D_0$ is the diffusion coefficient of CR in the ISM 

$$\frac{dn}{dt}=\frac{1}{r^2}\frac{d}{dr}(D(r)r^2\frac{dn}{dr})$$<br>
$$\frac{dn}{dt}=\frac{1}{r^2}\frac{d}{dr}(\frac{D_0}{(1+\frac{r}{r_c})^2}r^2\frac{dn}{dr})$$<br>
$$0=\frac{D_0r}{(1+\frac{r}{r_c})^3}[-\frac{2r}{r_c}\frac{dn}{dr}+2(1+\frac{r}{r_c})\frac{dn}{dr}+r(1+\frac{r}{r_c})\frac{d^2n}{dr^2}]$$<br>
Since $\frac{D_0r}{(1+\frac{r}{r_c})^3} \neq 0$, therefore the non-trivial solution comes from
$$0=-\frac{2r}{r_c}\frac{dn}{dr}+2(1+\frac{r}{r_c})\frac{dn}{dr}+r(1+\frac{r}{r_c})\frac{d^2n}{dr^2}$$<br>
$$\frac{d^2n}{dr^2}= \frac{-1}{r(1+\frac{r}{r_c})}[2(1+\frac{r}{r_c})-2\frac{r}{r_c}]$$

$D_0 = 3x10^{28}cm^2 s^{-1}$, $r_c = 10 pc$ 



```python
def dU_dr(U, r):
    # Here U is a vector such that n=U[0] and z=U[1]. This function hence returns [n', z']
    return [U[1], (-1/(r*(1+r/10)))*((2*(1+r/10))-(2*r/10))*U[1]]

U0 = [10**-9, 10**-9]
rs = np.arange(10**-5, 20, 0.001)
                 
Us = odeint(dU_dr, U0, rs)
ns = Us[:,0]

#plotting the solution. 
plt.xlabel("r (pc)")
plt.ylabel("n (cm^-3)")
plt.title("D(r)=$D_0/(1+r/r_c)^2$")
plt.plot(rs, ns)
plt.semilogy(rs, ns)
#plt.xlim([0, 0.1])
#plt.ylim([1.002e-9, 1.0035e-9])
plt.show()
```
